### Make a seperate column for rating variable and stars. ###

```python
import re
import numpy as np
def sep_col(text):
  if not isinstance(text, str) or text == 'Not rated yet':
    return np.nan
  else:
    rating = re.search(r"(\d+)(?=\s+ratings)", text)
    if rating:
      return rating.group(1)
    else:
      return np.nan
```

```python
df.insert(7,'Ratings',df['stars'].apply(sep_col))
```

```python
df
```

```python
import re
import numpy as np

def extract_stars(text):
    if not isinstance(text, str):
        return 0
    if text.strip().lower() == "not rated yet":
        return 0

    # pattern: "5 out of 5 stars" → pehla digit le lega
    match = re.search(r"(\d+)\s+out of 5 stars", text)
    if match:
        return int(match.group(1))
    return 0

df['Star_Rating'] = df['stars'].apply(extract_stars)
```

```python
df.drop(columns=['stars'],inplace=True)
df
```

### In the book names, there is '#', ',','[]' and 'Book' keyword present -> Consistenc ###


```python
def clean_book_name(name):
    if not isinstance(name, str):
        return name

    # 1. Remove text inside square brackets [ ... ]
    name = re.sub(r"\[.*?\]", "", name)

    # 2. Remove '#' symbols and numbers after them (e.g., #11, #12)
    name = re.sub(r"#\d+", "", name)

    # 3. Remove the word 'Book' followed by a number (e.g., Book 1, Book 2)
    name = re.sub(r"\bBook\s*\d*\b", "", name, flags=re.IGNORECASE)

    # 4. Remove extra commas
    name = name.replace(",", " ")

    # 5. Remove extra spaces
    name = re.sub(r"\s+", " ", name).strip()

    return name
```

```python
df['name'] = df['name'].apply(clean_book_name)
```

### Incorrect Datatype of release date, time, price. -> Accuracy ###

```python
df['releasedate'] = pd.to_datetime(df['releasedate'], format='%d-%m-%y')
```

```python
def convert_to_minutes(text):
    if not isinstance(text, str):
        return np.nan

    hours = re.search(r"(\d+)\s*hr", text)
    mins = re.search(r"(\d+)\s*min", text)

    total_mins = 0
    if hours:
        total_mins += int(hours.group(1)) * 60
    if mins:
        total_mins += int(mins.group(1))

    return total_mins
```

```python
df["time _duration"] = df['time'].apply(convert_to_minutes)
```

```python
df.drop(columns=['time'],inplace=True)
df
```

```python
df['price'] = df['price'].apply(lambda x:x.replace('Free','0'))
```

```python
df['price'] = df['price'].str.replace(',','')
df['price'] = df['price'].astype(int)
```

```python
df.rename(columns={'Ratings':'Ratings_count'},inplace=True)
```

```python
df['Ratings_count'] = df['Ratings_count'].astype(int)
```
